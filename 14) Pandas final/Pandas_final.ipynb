### <span style='background :yellow'>Statistica aggregata

#### <span style='background :lime'>Funzione GROUPBY( ) - caso 1 (grouped summaries)

```python
df_total = df.groupby('Type 1')['Total'].mean().sort_values(ascending=False).reset_index()
df_total.head()
```

```python
plt.figure(figsize=(14,6))

plt.barh(df_total['Type 1'], df_total['Total'])
plt.xlabel('Total')
plt.ylabel('Type 1')
plt.title('Total strenght per Type 1')

plt.xticks(ticks=np.arange(0,600,50))
plt.show()
```

#### <span style='background :lime'>Funzione GROUPBY( ) - caso 2 (multiple grouped summaries)

```python
df_total = df.groupby('Type 1')['Total'].agg([min, max, np.mean]).reset_index()
df_total.head()
```

```python
plt.figure(figsize=(14,6))

plt.barh(df_total['Type 1'], df_total['mean'], alpha = 0.5)
plt.barh(df_total['Type 1'], df_total['min'], alpha = 0.5)
plt.barh(df_total['Type 1'], df_total['max'], alpha = 0.5)
plt.xlabel('Total')
plt.ylabel('Type 1')
plt.title('Total strenght per Type 1')

plt.xticks(ticks=np.arange(0,850,50))
plt.show()
```

#### <span style='background :lime'>Funzione GROUPBY( ) - caso 3 (grouping by multiple variables)

```python
df_total = df.groupby(['Type 1', 'Type 2'])['Total'].mean().sort_values(ascending=False).reset_index()
df_total.head()
```

```python
df_total['Type'] = df_total['Type 1'] + ' - ' + df_total['Type 2']
df_total = df_total.drop(columns=['Type 1', 'Type 2'])
df_total = df_total[['Type', 'Total']] # ricreo il dataframe con le colonne spostate di indice
df_total.head()
```

#### <span style='background :lime'>Funzione GROUPBY( ) - caso 4 (many groups, many summaries)

```python
df_total = df.groupby(['Type 1', 'Type 2'])[['Total', 'Attack']].mean().sort_values(['Total','Attack'], ascending=False).reset_index()
df_total.head()
```

```python
df_total['Type'] = df_total['Type 1'] + ' - ' + df_total['Type 2']
df_total = df_total.drop(columns=['Type 1', 'Type 2'])
df_total = df_total[['Type', 'Total', 'Attack']] # ricreo il dataframe con le colonne spostate di indice
df_total.head()
```

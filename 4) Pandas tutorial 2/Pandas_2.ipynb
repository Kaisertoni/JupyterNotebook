### <span style='background :yellow'>Filtrare i dati

#### Maschere avanzate

##### Prima maschera
Filtra le righe per cui _Type 1_ = _Grass_ e _Type 2_ = _Fighting_

```python
advanced_mask_1 = (df['Type 1'] == 'Grass') & (df['Type 2'] == 'Fighting')
```

```python
df_filtered_1 = df.loc[advanced_mask_1] # è utile creare un nuovo dataframe in modo da avere sia quello filtrato che quello originale
df_filtered_1
```

##### Seconda maschera
Filtra le righe per cui (_Type 1_ = _Grass_ oppure _Type 2_ = _Fighting_) e _HP_ > 100

```python
advanced_mask_2 = ((df['Type 1'] == 'Grass') | (df['Type 2'] == 'Fighting')) & (df['HP'] > 100)
```

```python
df.loc[advanced_mask_2]
```

##### Terza maschera
Filtra le righe in cui in _Name_ è contenuta la parola Mega

```python
advanced_mask_3 = df['Name'].str.contains('Mega')
```

```python
df.loc[advanced_mask_3].head(5) # filtra tutte le righe in cui nel nome è contenuta la parola Mega
```

```python
df.loc[~advanced_mask_3].head(5) # filtra tutte le righe in cui nel nome non è contenuta la parola Mega
```

##### Quarta maschera (regexp)
Filtra le righe per cui _Type 1_ = _Fire_ o _Grass_

```python
import re
advanced_mask_4 = df['Type 1'].str.contains('fire|grass', flags=re.IGNORECASE, regex=True)
```

```python
df.loc[advanced_mask_4].head(5)
```

##### Quinta maschera (regexp)
Filtra le righe per cui _Name_ inzia con _Pi_<br>
__^__ indica deve iniziare con (quindi deve iniziare con pi)<br>
__.__ indica un qualsiasi carattere ad eccezione di una nuova linea<br>
__\*__ indica zero o più volte

```python
import re
advanced_mask_5 = df['Name'].str.contains('^pi.*', flags=re.IGNORECASE, regex=True)
```

```python
df.loc[advanced_mask_5].head(5)
```

### <span style='background :yellow'>Gestione di grossi dati

```python
df.head(10)
```

```python
new_df = pd.DataFrame()
```

```python
for x in pd.read_csv('./Dataset/pokemon_data.csv', chunksize=5):
    results = x.groupby('Type 1').mean()['HP']
    print("TTTTT",results,'\n\n\n\n')
    new_df = pd.concat([new_df,results])
    print('new_df:\n',new_df)
```
